```python
%matplotlib inline
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("VIGNESH121.xlsx")
df.head()
```

```python
d={"NO":100,"MR":80,"KG":60,"ST":45,"LT":35,"SM":25,"LO":17,"MT":15,"RL":13,"QM":11,"SH":9,"PR":7,"PK":6,"PD":5,"CM":4,"KL":3,"HN":2,"RM":1,"LB":0.5,"GM":0.25}
df['unit']=df['unit'].map(d)
df.head()
```

```python
features=['unit','price']
df1=df[features]
```

```python
df2=df[:1000]
df3=df1[:1000]
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df3)
```

```python
km=KMeans(n_clusters=400,init='k-means++',n_init=10,algorithm = 'auto',max_iter=400)
km.fit(X_scaled)
df2['cluster']=km.fit_predict(X_scaled)
df2[:20]
```
